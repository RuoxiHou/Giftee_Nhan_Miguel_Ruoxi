Data Cleaning for Nintendo and Lego Catalog

```python
import pandas as pd
import re
```

```python
lego_items = pd.read_csv('lego_catalog.csv')
lego_items.head()
```

```python
lego_items.duplicated().value_counts()
```

```python
lego_items.info()
```

```python
lego_items['age'] = lego_items['age'].replace('[^0-9]', '', regex=True)  # Remove non-numeric characters
lego_items['age'] = pd.to_numeric(lego_items['age'], errors='coerce')  # Convert to float, handle non-convertible to NaN
lego_items
```

```python
lego_items['price'] = lego_items['price'].str.replace('€', '')
lego_items['price'] = lego_items['price'].str.replace(',', '.')
```

```python
lego_items['price'] = lego_items['price'].replace('[^\d.]', '', regex=True)  # Remove non-numeric characters
# Convert to numeric, handle non-convertible to NaN
lego_items['price'] = pd.to_numeric(lego_items['price'], errors='coerce')
```

```python

display(lego_items.info())
lego_items.head()
```

```python
#Change column name of 'price' into 'suggested_price(€)'
lego_items.rename(columns={'price': 'suggested_price(€)'}, inplace=True)
```

```python
lego_items.head()
```

```python
lego_items.describe()
```

```python
#Find the records with age more than 18
lego_items[lego_items['age'] >18]
```

```python
#After checking the websites, they are actually from the first number to the last two number, for instance 599 means from 5 to 99
#Therefore, we only extract the first number of them
lego_items.loc[lego_items['age'] > 18, 'age'] = lego_items.loc[lego_items['age'] > 18, 'age'].astype(str).str.extract(r'(\d+)').astype(float)
lego_items[lego_items['age'] >18]
```

```python
lego_items.describe()
```

```python
nintendo_games = pd.read_csv('nintendo_game_catalog.csv')
nintendo_games.head()
```

```python
nintendo_games['price'] = nintendo_games['price'].str.replace(',', '.')
nintendo_games['price'] = nintendo_games['price'].replace('[^\d.]', '', regex=True)  # Remove non-numeric characters
# Convert to numeric, handle non-convertible to NaN
nintendo_games['price'] = pd.to_numeric(nintendo_games['price'], errors='coerce')
nintendo_games.head()
```

```python
#Change column name of 'price' into 'suggested_price(€)'
nintendo_games.rename(columns={'price': 'suggested_price(€)'}, inplace=True)
```

```python
#Save all the cleaned data in csv
lego_items.to_csv('lego_items.csv')
nintendo_games.to_csv('nintendo_games.csv')
```
