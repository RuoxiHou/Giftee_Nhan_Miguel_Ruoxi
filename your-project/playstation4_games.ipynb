```python
ps_games.drop(columns = ['discounted_price', 'discount', 'product_type', 'sale_name', 'date_recorded'], inplace = True)
```

```python
# According to the currency exchange rate on 05/12/2023, the rate is one pound equals to 1.17 Euro
# Therefore we exchange the pound into euro
# First we remove the currency mark
ps_games['original_price'] = ps_games['original_price'].replace('[^\d.]', '', regex=True)
```

```python
ps_games['original_price'] = pd.to_numeric(ps_games['original_price'], errors='coerce')
```

```python
# create a new column with the euro
ps_games['suggested_price(€)'] = [price*1.17 for price in ps_games['original_price']]
```

```python
ps_games.rename(columns={'name': 'title'}, inplace = True)
```
