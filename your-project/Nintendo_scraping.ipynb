```python
def first_5_pages(age_link):
    group_name = age_link.text.strip()
    if group_name not in age_group.keys():
        age_group[group_name] = [title_list, price_list, system_list, pic_list]

    for i in range(5):
        if i <= 1:
            current_page = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='VA_MainSearch']/div[2]/div/div/div[3]/div/button[position()=" + str(i + 1) + "]")))
        else:
            current_page = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='VA_MainSearch']/div[2]/div/div/div[3]/div/button[position()=" + str(i + 2) + "]")))
        time.sleep(1)
        current_page.click()
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        #Find all the items on one page
        group_items= soup.find_all('li', attrs={'class': "searchresult_row page-list-group-item col-xs-12"})[2:]
        #Find all the details from each item on one page 
        for item in group_items:
            time.sleep(1)
            img_tag = item.find('img')
            # Check if 'img' tag exists and has a 'src' attribute
            if img_tag and 'src' in img_tag.attrs:
                src_value = img_tag['src']
                pic_list.append(src_value)
            else:
                pic_list.append('None') 

            title_tag = item.find('p', class_='page-title')

            # Check if 'p' tag with class 'page-title' exists
            if title_tag:
                # Find the 'span' tag within the 'p' tag
                span_tag = title_tag.find('span', class_='page-title-text')

                # Check if 'span' tag exists
                if span_tag:
                    title = span_tag.get_text(strip=True)
                    title_list.append(title)
                else:
                    title_list.append('None') 
            else:
                title_list.append('None') 

            page_data_tag = item.find('p', class_='page-data')
            if page_data_tag:
                # Find the 'span' tag within the 'p' tag
                span_tag = page_data_tag.find('span')

                # Check if 'span' tag exists
                if span_tag:
                    system = span_tag.get_text(strip=True)
                    system_list.append(system)
                else:
                    # Handle the case where 'span' is not present
                    system_list.append('None')
            else:
                system_list.append('None')
            
            #Only get the original Price, and no discount is included
            price_small_tag = item.find('p', class_='price-small')
            if price_small_tag:
                # Find the 'span' tag within the 'p' tag
                original_span_tag = price_small_tag.find('span', class_='original-price')
                # Check if 'span' tag exists
                if original_span_tag:
                    price = original_span_tag.get_text(strip=True)
                    price_list.append(price)
                else:
                    span_tag = price_small_tag.find_all('span', class_='')[-1]
                    if span_tag:
                        price = span_tag.get_text(strip=True)
                        price_list.append(price)
                    else:
                        price_list.append('None')
            else:
                price_list.append('None')

        i = i +1
        time.sleep(1)
    return age_group    
```

```python
# Find the label with text "Leeftijd"
def fresh_age_links():
    leeftijd_label = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.XPATH, '//label[text()="Leeftijd"]'))
    )

    if leeftijd_label:
        # Scroll to the leeftijd_label to make it clickable
        actions = ActionChains(driver)
        actions.move_to_element(leeftijd_label).perform()

        # Click on the leeftijd_label using JavaScript
        driver.execute_script("arguments[0].click();", leeftijd_label)

        # Now you can find all the li elements under leeftijd_checkbox_list
        leeftijd_checkbox_list = leeftijd_label.find_element(By.XPATH, './following-sibling::ul')
        age_links = leeftijd_checkbox_list.find_elements(By.TAG_NAME, 'li')
    
    return age_links
```

```python

age_link_1 = fresh_age_links()[0]
age_link_1 = WebDriverWait(driver, 10).until(EC.visibility_of(age_link_1))
age_link_1.click()

first_5_pages(age_link=age_link_1)

age_link_2 = fresh_age_links()[1]
age_link_2 = WebDriverWait(driver, 10).until(EC.visibility_of(age_link_2))
age_link_2.click()

first_5_pages(age_link=age_link_2)

age_link_3 = fresh_age_links()[2]
age_link_3 = WebDriverWait(driver, 10).until(EC.visibility_of(age_link_3))
age_link_3.click()

first_5_pages(age_link=age_link_3)

age_link_4 = fresh_age_links()[3]
age_link_4 = WebDriverWait(driver, 10).until(EC.visibility_of(age_link_4))
age_link_4.click()

first_5_pages(age_link=age_link_4)

age_group
```
